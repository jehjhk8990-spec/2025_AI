```python
import torch

#3*3 텐서 생성
x = torch.Tensor([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
print(x)
print("Shape of tensor:", x.shape) # torch.Size([2, 3])
print("Type of tensor:", x.dtype) # torch.float32 (기본값)
print("Allocated device:", x.device) # cpu 또는 cuda
print("랭크(차원):", x.ndimension())
```

```python
1# 텐서 다루기 (모양 확인)
x = torch.Tensor([[1, 2, 3], [4, 5, 6], [7, 8, 9]])

# 특정 차원에 대해 1 랭크 올리기
x = torch.unsqueeze(x, 0)
print(x.shape)

# 1 크기의 차원을 가지는 랭크 없애기
x = torch.squeeze(x)
print(x.shape)

# 텐서 모양 변경
x = x.view(9)
print(x.shape)
```

```python
# 1. 크기만 지정해서 비어있는 텐서 생성
empty_tensor = torch.empty(3, 3)
print("empty_tensor:", empty_tensor)

# 2. 0으로 채워진 텐서 생성
zeros_tensor = torch.zeros(3, 3)
print("zeros_tensor:", zeros_tensor)

# 3. 1로 채워진 텐서 생성
ones_tensor = torch.ones(3, 3)
print("ones_tensor:", ones_tensor)

# 4. 랜덤 값으로 채워진 텐서 생성
rand_tensor = torch.rand(3, 3)
print("rand_tensor:", rand_tensor)
```

```python
import torch
import torch.nn as nn

# 모델 정의 (2층 FC 신경망)
class simple_model(nn.Module):
    def _init_(self, input_size, hidden_size, num_classes):
       super(simple_model, self)._init_()
       self.fc1 = nn.Linear(input_size, hidden_size)
       self.relu = nn.Relu()
       self.fc2 = nn.Linear(hidden_size, num_classes)

   def forward(self, x):
       out = self.fc1(x)
       out = self.relu(out)
       out = self.fc2(out)
       return out
```
