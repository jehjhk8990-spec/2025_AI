```python
import matplotlib.pyplot as plt
import numpy as np
path = '/content/drive/MyDrive/2025_AI_Education/seoul_temp.txt'
data = np.loadtxt(path, dtype='str')
```

```python
y = data[:, 2].astype(float) # y축 설정
x = range(len(y)) # x축 정의
plt.plot(x, y) # 그림 그리기
plt.xlabel('day') # x축 라벨 설정
plt.ylabel('average temperature') # y축 라벨 설정
plt.title('Average Temperature in Seoul') # 그림 제목 설정
plt.show() # 그림 출력
```

```python
N = 100
dx = 2*np.pi/N
for i in range(N):
  x = i*dx
  y = np.sin(x)
```

```python
N = 100
dx = 2*np.pi/N
x_arr = []
y_arr = []
for i in range(N):
  x = i*dx
  y = np.sin(x)
  x_arr.append(x)
  y_arr.append(y)
plt.plot(x_arr, y_arr, 'o')
plt.plot(x_arr, y_arr)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
x = np.random.normal(0, 1, 10) #평균 0, 표준편차 1인 정규분포에서 10개를 뽑는다.
y = np.random.normal(0, 1, 10)
sizes = np.random.uniform(10, 100, 10) # 10과 100 사이의 10개를 크기로 뽑는다.
colors = np.random.uniform(0, 1, 10) # 0에서 1 사이의 값으로 색을 표시한다.
plt.scatter(x, y, s=sizes, c=colors)
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar() # 색으로 표시된 값의 범위를 그래프의 우측에
plt.show
```

```python
x = np.arange(7) + 0.5 # 0에서 6까지 정수에 0.5를 더한다.
y = np.random.uniform(1, 9, len(x)) # 1에서 9 사이의 난수
plt.bar(x, y, width=1, edgecolor='white') # 막대그래프의 경계를 흰색으로 한다.
plt.xlabel("days")
plt.ylabel("hours")
plt.show
```

```python
x = np.random.normal(0, 1, 10000) # 평균이 0, 표준편차가 1인 정규분 # 포 난수 10000개
plt.hist(x, bins=15, edgecolor='white') # 도수분포를 15개를 구간으로 나눈다. plt.xlabel("x")
plt.ylabel("frequency")
plt.show()
```
